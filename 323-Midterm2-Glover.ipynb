```python
def coeff(n):
    
    if n == 0:
        return 0
    else:
        fact = 2/(n*pi)
        term1 = cos((n*pi)/2)
        term2 = (-1)**n

    return fact*( term1 - term2 )
```

```python
def n_eigenstate(n,x):
    fact = sqrt(2/L)
    term1 = sin((n*pi*x)/(L))
    return fact*term1
```

```python
t = arange(0,1000)
```

```python
%matplotlib inline  
```

```python
L = 1
x = linspace(0,L,500)
top = 25
n = arange(0,top)

wavefunc = 0
for i in range(0,len(n)+1):
    wavefunc = wavefunc + (coeff(i))*(n_eigenstate(i,x))


figure(figsize=(10,6), dpi= 100)
plot(x,wavefunc,label='Approximation after {} terms'.format(top))
xlim(-0.1,L+0.1);ylim(-0.3,2.5)

#The ends, middle of box, and bottom of box
arrow(0,0,0,2,head_width=0.025,fill=0);arrow(L,0,0,2,head_width=0.025,fill=0);vlines(0.5,0,2,linestyle='--')
hlines(0,0,L)

#horizontal lines which make up the exact wavefunction
hlines(0,0,L/2,color='orange'); hlines(sqrt(2/L),L/2,L,color='orange',label='Wave function')

legend(bbox_to_anchor=[0.5, 1])

yticks([])
```

```python
def init(): #init for initial frame of the animation

    L = 1
    x = linspace(0,L,100)
    top = 0
    n = arange(0,top)

#     wavefunc = 0
#     for i in range(0,len(n)+1):
#         wavefunc = wavefunc + (coeff(i))*(n_eigenstate(i,x))
    xlim(-0.1,L+0.1);ylim(-0.3,2.5)

    #The ends, middle of box, and bottom of box
    arrow(0,0,0,2,head_width=0.025,fill=0);arrow(L,0,0,2,head_width=0.025,fill=0);vlines(0.5,0,2,linestyle='--')
    hlines(0,0,L)

    #horizontal lines which make up the exact wavefunction
    hlines(0,0,L/2,color='orange'); hlines(sqrt(2/L),L/2,L,color='orange',label='Wave function')

    legend(bbox_to_anchor=[0.5, 1])
    yticks([])
    xlabel('$x/L$')
    
    return fig,

def animate(i): #something that changes frame to frame
    #animate is a function of i which  is the frame number 
    fig.clear()
    x = linspace(0,L,100)
    top = i*2
    n = arange(0,top)

    wavefunc = 0
    for val in range(0,len(n)+1):
        wavefunc = wavefunc + (coeff(val))*(n_eigenstate(val,x))
    plot(x,wavefunc,color='blue',label='Approximation')
    
    
    xlim(-0.1,L+0.1);ylim(-0.3,2.5)

    #The ends, middle of box, and bottom of box
    arrow(0,0,0,2,head_width=0.025,fill=0);arrow(L,0,0,2,head_width=0.025,fill=0);vlines(0.5,0,2,linestyle='--')
    hlines(0,0,L)

    #horizontal lines which make up the exact wavefunction
    hlines(0,0,L/2,color='orange'); hlines(sqrt(2/L),L/2,L,color='orange',label='Wave function')

    legend(bbox_to_anchor=[0.5, 1])
    yticks([])
    xlabel('$x/L$')
   
    title('Approximation with {} terms'.format(top))
    return fig,
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ani = animation.FuncAnimation(fig, animate, init_func=init,frames=100, interval=50)
ani.save('3D.gif',writer='pillow',fps=10)
plt.close()
```
