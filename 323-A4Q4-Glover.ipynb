where $$c_n = \int_0^2 \psi^* e_n dx  = \int_0^2 \frac{1}{2} e_n dx $$

```python
def e0(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    return (1/sqrt(2)) * (fact*exponz*expont)
```

```python
def e1(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = z*sqrt(2)
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
def e2(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (2*z**2 -1)/sqrt(2)
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
def e3(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (z*(2*z**2 -3))/sqrt(3)
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
def e4(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (4*z**4 -12*z**2 +3)/(2*sqrt(6))
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
def e5(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (z*(4*z**4 -20*z**2 +15))/(2*sqrt(15))
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
def e6(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (8*z**6 -60*z**4 +90*z**2 -15)/(12*sqrt(5))
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
def e7(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (z*(8*z**6 -84*z**4 +210*z**2 -105))/(6*sqrt(70))
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
def e8(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (16*z**8 -224*z**6 + 840*z**4 -840*z**2 +105)/(24*sqrt(70))
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
def e9(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (z*(16*z**8 -288*z**6 + 1512*z**4 -2520*z**2 +945))/(72*sqrt(35))
    return (1/sqrt(2))*fact*exponz*expont*term1
```

```python
c0 = quad(e0,0,2,args=(0))[0]
c1 = quad(e1,0,2,args=(0))[0]
c2 = quad(e2,0,2,args=(0))[0]
c3 = quad(e3,0,2,args=(0))[0]
c4 = quad(e4,0,2,args=(0))[0]
c5 = quad(e5,0,2,args=(0))[0]
c6 = quad(e6,0,2,args=(0))[0]
c7 = quad(e7,0,2,args=(0))[0]
c8 = quad(e8,0,2,args=(0))[0]
c9 = quad(e9,0,2,args=(0))[0]

c = [c0,c1,c2,c3,c4,c5,c6,c7,c8,c9]
for i,val in enumerate(c):
    print('The c{} coefficient is {:.4f}'.format(i,val))
sum(c)
```

The wave function is an infinite sum since there are an infinite number of energy levels for the SHO, so these first 10 terms only represent a small fraction of the total number of terms.  However taking the appropriate number of terms can approximate the actual wave function very well

```python
def e0_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    return  (fact*exponz*expont)

def e1_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = z*sqrt(2)
    return fact*exponz*expont*term1

def e2_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (2*z**2 -1)/sqrt(2)
    return fact*exponz*expont*term1

def e3_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (z*(2*z -3))/sqrt(3)
    return fact*exponz*expont*term1

def e4_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (4*z**4 -12*z**2 +3)/(2*sqrt(6))
    return fact*exponz*expont*term1

def e5_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = z*(4*z**4 -20*z**2 +15)/(2*sqrt(15))
    return fact*exponz*expont*term1

def e6_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (8*z**6 -60*z**4 +90*z**2 -15)/(12*sqrt(5))
    return fact*exponz*expont*term1

def e7_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (z*(8*z**6 -84*z**4 +210*z**2 -105))/(6*sqrt(70))
    return fact*exponz*expont*term1

def e8_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (16*z**8 -224*z**6 + 840*z**4 -840*z**2 +105)/(24*sqrt(70))
    return fact*exponz*expont*term1

def e9_factor(z,t):
    fact = 1/(pi**(1/4))
    exponz = e**(-z**2 / 2)
    expont = e**(-t/2)
    term1 = (z*(16*z**8 -228*z**6 + 1512*z**4 -2520*z**2 +945))/(72*sqrt(35))
    return fact*exponz*expont*term1
```

```python
def psi_prob(z,t):
    term0 = c0* e0_factor(z,t)
    term1 = c1* e1_factor(z,t)
    term2 = c2* e2_factor(z,t)
    term3 = c3* e3_factor(z,t)
    term4 = c4* e4_factor(z,t)
    term5 = c5* e5_factor(z,t)
    term6 = c6* e6_factor(z,t)
    term7 = c7* e7_factor(z,t)
    term8 = c8* e8_factor(z,t)
    term9 = c9* e9_factor(z,t)
    return (term0+term1+term2+term3+term4+term5+term6+term7+term8+term9)**2
```
