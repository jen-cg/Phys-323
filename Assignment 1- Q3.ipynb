$4.5e^{i3.4} + 2.1e^{i1.4}$

Using either a ruler and protractor or a computer (e.g. Matlab, Python, etc), plot the
above addition operation on a real-imaginary plane, along with the result. Take the x-axis to
be the real-axis, and the y-axis to be the imaginary-axis.


let $a = 4.5e^{i3.4}$ and $ b = 2.1e^{i1.4}$

In cartesian form $ a = 4.5cos(3.4) + i4.5sin(3.4)$

and $b = 2.1cos(1.4) + i2.1sin(1.4)$

i can be thought of as a sort of unit vector along the imaginary axis.
Then each complex number becomes a vector of real and imaginary components.


```python
%pylab nbagg

from matplotlib.pyplot import *
%matplotlib inline

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

```python
# number = [ real component , imaginary component ]

a =  [ 4.5*cos(3.4*pi/180) , 4.5*sin(3.4*pi/180) ] #must convert degrees to radians, hence pi/180
b = [ 2.1*cos(1.4*pi/180) , 2.1*sin(1.4*pi/180) ]

aplusb = [ a[0]+b[0] , a[1]+b[1] ] # [real+real,imaginary+imaginary] 
```

```python
figure(1)
plot( [0,a[0]],[0,a[1]], color='g',label='Vector a')
plot( [0,b[0]],[0,b[1]], color='b',label='Vector b')
plot( [0,aplusb[0]],[0,aplusb[1]], color='r',label='Vector a plus  b')
xlabel('Real axis');ylabel('Imaginary axis')
legend()
grid()
```
