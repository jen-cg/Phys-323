```python
from scipy import optimize
```

```python
def even( z  ):
    z0 = 3
    return z*tan(z) - sqrt( z0**2 - z**2 )
```

```python
def odd( z ):
    z0 = 3
    return z*(cos(z)/sin(z)) + sqrt( z0**2 - z**2 )
```

```python
optimize.fsolve(even,x0=1)
```

```python
optimize.fsolve(odd,x0=2)
```

```python
(optimize.newton(even,x0=1,maxiter=100))**2
```

```python
(optimize.newton(odd,x0=2,maxiter=100))**2
```

```python
z = linspace(0,2.9999,1000)
z2 = linspace(0,1.5,100)
z3 = linspace(0,2.6,100)
```

```python
close(1);figure(1)
plot(z,sqrt(3**2-z**2),label='$z\sqrt{9 - z^2}$')
plot(z2,z2*tan(z2),label='$z tan(z)$')
plot(z3,-z3*(cos(z3)/sin(z3)),label='$-z cot(z)$')
ylim(0,3.5)
vlines(1.17,0,3.5,linestyle='--'); vlines(2.27,0,3.5,linestyle='--')
legend()
grid()
xlabel('Z')
ylabel('f(z)')
```
