```python
m = 9.1093837e-31
hbar = 1.05457182e-34
w = 2*pi*(5e15)
```

```python
def psi0 (x ):
    alpha = (m*w)/(hbar)
    y = sqrt(alpha) * x
    
    term1 = (alpha / pi)**(1/4)
    term2 = e**( - (y**2) / 2 )
    
    return term1*term2
```

```python
def psi1(x):
    alpha = (m*w)/(hbar)
    y = sqrt(alpha) * x
    
    term1 = (alpha / pi)**(1/4)
    term2 = e**( - (y**2) / 2 )
    term3 = sqrt(2)*y
    
    return term1*term2*term3
```

```python
def psi2(x):
    alpha = (m*w)/(hbar)
    y = sqrt(alpha) * x
    
    term1 = (alpha / pi)**(1/4)
    term2 = e**( - (y**2) / 2 )
    term3 = 1/sqrt(2)
    term4 = (2*y**2 - 1)
    
    return term1*term2*term3*term4
```

```python
def E(n):
    return hbar*w*(n + 1/2)
```

```python
x = linspace(-2e-10 , 2e-10,1000)
```

```python
# close(1);figure(1)
figure(figsize=(8,8),dpi= 100)
i = 5
plot(x , i*E(0) + (10e-6)*E(2)*psi0(x) ,color='lightblue',label='Ground State')
hlines(i*E(0),-2e-10, 2e-10,color='lightblue',linestyle='--',label='$E_0$')
fill(x,i*E(0) + (10e-6)*E(2)*psi0(x),color='lightblue',alpha=0.25)

plot(x , i*E(1) + (10e-6)*E(2)*psi1(x),color='pink',label='First Excited State')
hlines(i*E(1),-2e-10, 2e-10,color='pink',linestyle='--',label='$E_1$')
fill(x,i*E(1) + (10e-6)*E(2)*psi1(x),color='pink',alpha=0.25)


plot(x , i*E(2) + (10e-6)*E(2)*psi2(x),color='orange',label='Second Excited State')
hlines(i*E(2),-2e-10, 2e-10,color='orange',linestyle='--',label='$E_2$')
fill(x,i*E(2)+ (10e-6)*E(2)*psi2(x) ,color='orange',alpha=0.25)

legend(bbox_to_anchor=[1,-0.1],ncol=3)

xlabel('Position ( x )')
ylabel('Potenrial ( J )')
```
