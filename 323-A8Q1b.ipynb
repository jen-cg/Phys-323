```python
def fact( n_x , n_y , n_z ):
    return (1/12)*( 12*n_x + 4*n_y + 6*n_z + 11 )
```

```python
# n_x , n_y , n_z can take on any interger value 0, 1, 2, 3 to infinity 
top = 50 #"top" is the value at which we choose to trucate this list
n_x = arange(0,top,1)
n_y = arange(0,top,1)
n_z = arange(0,top,1)
```

```python
combos = {} 
factors = [] 
#This will go through every possible combination of n_x , n_y , n_z 
for x in n_x:
    for y in n_y:
        for z in n_z:
            val = fact(x,y,z)
            if val not in factors:
                combos[val] = [ (x,y,z) ]   
            else:
                combos[val].append((x,y,z))
            factors.append( val )
            
```

```python
print('The ten lowest energy levels with their corresponding degeneracy and the combinations of (n_x, n_y, n_z) eigenvalues which produce them are as follows \n')
print('Factor | Degeneracy |   (n_x, n_y, n_z) \n----------------------------------')
for i , key in enumerate(sorted(combos.keys())):
    #Iterating through each possible factor from lowest to highest
    if i < 10: #The question only asks for the first ten distinct states 
        print('{:.3f}    |     {}      |   {}'.format(key , factors.count(key) , combos[key]  ))
```

Note: "factor" is the coefficient $c$ of $E = c \hbar \omega_0$
